# Supervised Learning

```python
X1 = df3[['Wage', 'Value']]
y1 = df3['International Reputation']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X1, y1, test_size=0.2, random_state=20)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='poly')
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# Visualisasi Confusion Matrix dengan Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap="Blues")
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Confusion Matrix Multiclass')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## Hasil Plot

```python
from sklearn.inspection import DecisionBoundaryDisplay as dbd

plt.figure(figsize = (10,8))
sns.scatterplot(x='Wage', y='Value', hue='International Reputation', data=df3)
plt.axis([0.13, 1.2, 0.13, 1.2])

ax = plt.gca()

dbd.from_estimator(
    estimator=svm,
    X=X1,
    plot_method="contourf",
    alpha=0.25,
    linestyles=["--", "-", "--"],
    ax=ax,
)

ax.scatter(
    svm.support_vectors_[:, 0],
    svm.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    alpha=0.25,
    facecolors="none",
    edgecolors="k",
)

plt.xlabel('Wages')
plt.ylabel('Values')
plt.show()
```

# Unsupervised Learning

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score, homogeneity_score, completeness_score
```

```python
X = df[['Overall', 'Potential']]
```

```python
# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## KMEANS

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

## Plot Elbow

```python
# Membuat plot Elbow
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Metode Elbow untuk Menentukan Jumlah Klaster')
plt.xlabel('Jumlah Klaster')
plt.ylabel('WCSS')
plt.show()
```

## Melatih model KMEANS

```python
# Berdasarkan plot Elbow, pilih jumlah klaster yang sesuai
num_clusters = 4

# Melatih model K-Means dengan jumlah klaster yang telah ditentukan
kmeans = KMeans(n_clusters=num_clusters, init='k-means++', random_state=42)
kmeans.fit(X_scaled)
df['Cluster'] = kmeans.labels_
```

## Evaluasi klaster tanpa label asli dataset

```python
silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
calinski_harabasz = calinski_harabasz_score(X_scaled, kmeans.labels_)
davies_bouldin = davies_bouldin_score(X_scaled, kmeans.labels_)

print(f'Silhouette Score: {silhouette_avg}')
print(f'Calinski-Harabasz Score: {calinski_harabasz}')
print(f'Davies-Bouldin Score: {davies_bouldin}')
```

```python
# Visualisasi hasil klaster
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.title('Hasil K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

## Evaluasi klaster menggunakan label asli Dataset

```python
real_labels = df['Cluster']  
predicted_labels = kmeans.labels_

ari = adjusted_rand_score(real_labels, predicted_labels)
ami = adjusted_mutual_info_score(real_labels, predicted_labels)
homogeneity = homogeneity_score(real_labels, predicted_labels)
completeness = completeness_score(real_labels, predicted_labels)

print(f'Adjusted Rand Index (ARI): {ari}')
print(f'Adjusted Mutual Information (AMI): {ami}')
print(f'Homogeneity: {homogeneity}')
print(f'Completeness: {completeness}')

# Visualisasi hasil klaster dengan pusat klaster
plt.figure(figsize=(10, 8))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.title('Hasil K-Means Clustering')
plt.xlabel('Fitur 1 (Overall)')
plt.ylabel('Fitur 2 (Potential)')
plt.legend()
plt.show()
```
