### Mengimport library dan juga mendeklarasikan file csv

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
```

```python
df = pd.read_csv('FIFA 19 data.csv')
df
```

### Bagi dataset menjadi training set dan testing set dengan proporsi 70:30
`codingan:`

```python
X = df.drop('Overall', axis=1) # variabel independen 
y = df['Overall'] # variabel dependen 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # membagi data secara acak
```

### Normalisasi data pada salah satu attribute menggunakan Min Max scaler 
`codingan:`

```python
df_norm = df.copy() # membuat salinan dataset 
scaler = MinMaxScaler() # membuat objek scaler 
df_norm['Age'] = scaler.fit_transform(df_norm[['Age']]) # melakukan normalisasi pada kolom Age
```

### Standarisasi pada dataset 
`codingan:`

```python
df_std = df.copy() # membuat salinan dataset 
scaler = StandardScaler() # membuat objek scaler 
df_std = scaler.fit_transform(df_std[['Value', 'Wage']])# melakukan standarisasi 
```

### Data cleaning pada nilai null dan mengganti nilai null sesuai keinginan
`codingan:`

```python
df_clean = df.copy() # membuat salinan dataset 
df_clean.isnull().sum() # mengecek jumlah nilai null pada setiap kolom
```

### Dikarenakan tidak ada nilai null maka kita menambakan nilai null secara acak
`codingan:`

```python
np.random.seed(42) # menetapkan seed untuk reproduksi hasil 
df_clean.iloc[np.random.randint(0, len(df_clean), 10), np.random.randint(0, len(df_clean.columns), 10)] = np.nan # mengisi 10 nilai acak dengan null 
df_clean.isnull().sum() # mengecek kembali jumlah nilai null pada setiap kolom
```

### Mengganti nilai null sesuai ketentuan
`codingan:`

```python
df_clean['Age'].fillna(df_clean['Age'].median(), inplace=True) # mengganti nilai null pada kolom Age dengan median 
df_clean['Nationality'].fillna(df_clean['Nationality'].mode()[0], inplace=True) # mengganti nilai null pada kolom Nationality dengan modus 
```

### Data cleaning pada nilai duplikat
`codingan:`

```python
df_dup = df.copy() # membuat salinan dataset 
df_dup.duplicated().sum() # mengecek jumlah baris yang duplikat
```
