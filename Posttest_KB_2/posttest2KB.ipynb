### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.
`codingan:`

```python
# Mendapatkan informasi yang diminta
jumlah_record = len(df)
jumlah_attribute = len(df.columns)
nama_attribute = df.columns.tolist()
jumlah_attribute_angka = len(df.select_dtypes(include=['number']).columns)

# Membuat dictionary dengan informasi yang diinginkan
informasi = {
    'Jumlah Record': jumlah_record,
    'Jumlah Atribut': jumlah_attribute,
    'Nama Atribut': nama_attribute,
    'Jumlah Atribut dengan Tipe Angka': jumlah_attribute_angka
}

# Menampilkan isi dictionary
for key, value in informasi.items():
    print(f'{key}: {value}')
```
