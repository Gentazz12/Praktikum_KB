### Mengimport library dan juga mendeklarasikan file csv
`codingan:`

```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
df = pd.read_csv('FIFA 19 data.csv')
df.head()
```

### Mengubah Index angka menjadi nama pemain
`codingan:`

```python
df.set_index('Name', inplace=True)
df.head()
```

### Menampilkan Visualisasi Bar Plot
`codingan:`

```python
jumlah_pemain = df['Offensive Work Rate'].value_counts()
jumlah_pemain.plot.bar(ylabel = 'Jumlah Pemain', xlabel = 'Offensive Work Rate')
plt.title('Perbandingan Jumlah Pemain Berdasarkan Offensive Work Rate')
plt.xticks(rotation=60, ha='right')
```

Kesimpulan: kebanyakan pemain di game FIFA 19 memiliki Medium OWR

### Menampilkan Visual Histogram
`codingan:`

```python
df['Overall'].plot.hist(ylabel = 'Jumlah Pemain', xlabel = 'Overall Rate Pemain')
plt.title('Perbandingan Jumlah Pemain Berdasarkan Overall Pemain')
```

Kesimpulan: Kebanyakan pemain yang ada di game FIFA 19 memiliki overal rating rata rata 70

### Menampilkan Visual Pie chart
`codingan:`

```python
owr_group = df.groupby('Offensive Work Rate').sum()
explode = [0, 0, 0] 
owr_group['Overall'].sort_values().plot.pie(autopct='%.2f%%', startangle=90, explode=explode)
plt.title('Presentase Offensive Work Rate Berdasarkan Overall', fontsize = 25)
```

Kesimpulan: Berdasarkan Overall pemain di game FIFA 19 pemain yang memiliki OWR medium adalah 66,61%

```python
year_total_corr = df.loc[:, 'Value':'International Reputation'].corr()
year_total_corr
```

### Menampilkan Visual Heatmap
`codingan:`

```python
sns.heatmap(
    data=year_total_corr,
    annot=True
)
plt.title('Korelasi antar Value, Wage, Special, International Reputation', fontsize = 25)
```

Kesimpulan: Korelasi International Reputation dan Value memiliki nilai tinggi
